# Task1

## Write Sales Records to a File

```python
sales = [1200, 450, 980, 1500, 3000]

with open("sales_data.txt", "w") as file:
    for sale in sales:
        file.write(str(sale) + "\n")
    file.close()

with open("sales_data.txt", "r") as file:
    for sale in file:
        print(sale)
```

# Task 3

## append the lines

```python
append_list =[5000,2500,1700] 

with open("sales_data.txt", "a") as file:
    for sale in append_list:
        file.write(str(sale) + "\n")
    file.close()

total_files_lines = 0
with open("sales_data.txt", "r") as file:
    for sale in file:
        print(sale)
        total_files_lines += 1
    file.close()

print(f"Total number of lines in file: {total_files_lines}")    
```

# task 4

## generate summary

```python
all_sales_list = [int(sale) for sale in open("sales_data.txt", "r").readlines()]
print(f"all_sales_list: {all_sales_list}")
print(f"Total sales: {sum(all_sales_list)}")
print(f"Highest sales: {max(all_sales_list)}")
print(f"Lowest sales: {min(all_sales_list)}")
print(f"Average sales: {sum(all_sales_list) / len(all_sales_list)}")
```

# Task 5: Create Product Info File (User Input)

```python
first_product_name = input("Enter first product name: ")
fisrt_product_price = float(input("Enter first product price: "))

second_product_name = input("Enter second product name: ")
second_product_price = float(input("Enter second product price: "))

third_product_name = input("Enter third product name: ")
third_product_price = float(input("Enter third product price: "))

with open("products.txt", "w") as file:
    file.write(f"{first_product_name}|{fisrt_product_price}\n")
    file.write(f"{second_product_name}|{second_product_price}\n")
    file.write(f"{third_product_name}|{third_product_price}\n")
file.close()

with open("products.txt", "r") as file:
    for line in file:
        print(line)
    file.close()
```

# Task 7: Mini project - export discounted proces

```python
prices = {
    "Mouse": 500,
    "Keyboard": 800,
    "Monitor": 7000,
    "Pendrive": 400,
    "Camera": 5000
}

write_file = open("discounted_report.txt", "w");

discounted_percentages = []

for item, price in prices.items():
    discount_percentage = int(input(f"Enter discount percentage for {item}: "))
    discounted_percentages.append(discount_percentage)
    discount_amount = price * (discount_percentage / 100)
    final_price = price - discount_amount
    #print(f"{item}: {final_price}")
    write_file.write(f"{item}|{price}|{final_price}\n")

write_file.close()

#print(open("discounted_report.txt", "r").read())

with open("discounted_report.txt", "r") as file:
    lines = file.readlines()
    for line in lines:
        print(line)
    
    print(f"Total itesm: {len(lines)}")
    print(f"Average Discounted percentages: {sum(discounted_percentages)/len(discounted_percentages)}")
    
file.close()
```
