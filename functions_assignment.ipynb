```python
def apply_discount(price, discount_percent=5):
    if discount_percent > 60:
        discount_percent = 60
    return price * (100 - discount_percent)/100
    
print(apply_discount(1000,10))
print(apply_discount(500))
```

# Task2

```python
def factorial(n):
    if n == 0:
        return 1
    elif n < 0:
        return 'undefined'
    else:
        return n * factorial(n-1)

print(factorial(5))
print(factorial(0))
print(factorial(-3))
```

```python
prices = [100,250,400,1200,50]
prices_with_gst = list(map(lambda price: price * 1.18, prices))
print(prices)
print(prices_with_gst)
```

# Task 5 - Using filter(): Filter Expensive Products

```python
prices = [100, 250, 400, 1200, 50, 2000, 850]
#price_gt_500 = [price for price in prices if price > 500]
price_gt_500 = list(filter(lambda price: price > 500, prices))
print(price_gt_500)
price_lte_500 = list(filter(lambda price: price <= 500, prices))
print(price_lte_500)
```

# Task 6 - Combined Utility Function

```python
def process_prices(prices):
    discount_10_percent = list(map(lambda price: price * 0.9, prices))
    price_gt_300 = list(filter(lambda price: price > 300, prices))
    return discount_10_percent, price_gt_300

print(process_prices([100, 500, 900, 50, 750]))
```

# Task 7 - Mini Problem: Menu Using Functions

```python
#prices_list = []
def add_price(prices_list, price):
    prices_list.append(price)
    return prices_list

def get_average_price(prices_list):
    return sum(prices_list) / len(prices_list)

def get_max_price(prices_list):
    return max(prices_list)

# create menu Then create a simple menu:
# 1 → Add price
# 2 → Show average price
# 3 → Show highest price 
# q → Quit

prices_list = []

while True:
    print("\n=== MENU ===")
    print("1. Add price")
    print("2. Show average price")
    print("3. Show highest price")
    print("q. Quit")
    
    choice = input("Enter your choice (1/2/3/q): ").lower().strip()

    if choice == 'q':
        break
    elif choice == '1':
        add_price(prices_list, int(input("Enter price: ")))
    elif choice == '2':
        print(get_average_price(prices_list))
    elif choice == '3':
        print(get_max_price(prices_list))
```
