# Task1: safe division utility

```python
try:
    numerator = int(input("Enter numerator: "))
    denominator = int(input("Enter denominator: "))
    result = numerator / denominator
except ZeroDivisionError:
    print("Cannot divide by zero")
except ValueError:
    print("Invalid input. Please enter integers only.")
else:
    print("Result:", result)
finally:
    print("Operation completed.")
```

# Task 2: Bill calculator with Error Handling

```python
prices = [120, 350, 'abc', 500, -200, 800]
running_total = 0
for items in prices:
    try:
        if items < 0:
            raise ValueError("Negative value")
        running_total+=items
    except TypeError:
        print("Cannot have string in price")
    except ValueError:
        print("Negative price not allowed")
    finally:
        print("Running total:", running_total)
```

# Task 3: Custom Exception: Age Validator

```python
def check_age(age):
    if age < 1 or age > 120:
        raise ValueError("Age must be between 1 and 120 ")
    else:
        print(f"Age {age} is valid")

try:
    try:
        age = int(input("Enter age: "))
    except ValueError as e:
        print('Age should be numeric')
    else:
        check_age(age)
except ValueError as e:
    print(e)
```

# Task 4: File reader with exception handling

```python
file_name = input("Enter file name: ")
try:
    file = open(file_name, 'r')
except FileNotFoundError:
    print("File not found")
else:
    for i in range(3):
        print(file.readline())
    file.close()
finally:
    print("File operation attempted")
```
